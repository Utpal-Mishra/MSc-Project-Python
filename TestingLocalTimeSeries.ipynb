## EXPLORATORY DATA ANALYSIS

### Renaming Column Names

```
data.rename({"Date_HPSC": "Date"}, axis = 1, inplace = True)
data.head()
```

### Checking Empty Values

```
if data.isna().sum().sum() ==0:
  print("NO EMPTY VALUES!!!")
```

## Reframing Columns to Categorical Rows : TESTING Location

```
hospitalrecords = data[["Date", "Hospitals", "TotalLabs",	"NonHospitals"]] 
hospitalrecords = hospitalrecords[['Date', 'Hospitals', 'NonHospitals', 'TotalLabs']]
hospitalrecords.head()
```

```
hospitalrecords = hospitalrecords.melt(id_vars=["Date"], var_name="TestingLocation", value_name="CasesCount")
hospitalrecords.sort_values(["Date"], inplace = True)
hospitalrecords.head()
```

### VISUALIZATION

```
!pip install raceplotly
```

```
fig = px.bar(hospitalrecords, x='TestingLocation', y="CasesCount", animation_frame = "Date", color = "TestingLocation")
fig.show()
```

```
fig = px.area(hospitalrecords, x = "Date", y="CasesCount", color="TestingLocation", line_group="TestingLocation")
fig.show()
```
