### IMPORT LIBRARIES

```
import pandas as pd
import numpy as np
import plotly.express as px
from datetime import datetime as dt
from sklearn import preprocessing

print("LIBARARIES IMPORTED!!!")
```

### IMPORT DATA

```
path = "/content/drive/My Drive/Final Project/euro-exchange-rates.csv"

data = pd.read_csv(path)
print("Dimensions of Data: ", data.shape)
```

```
data.head()
```

## EXPLORATORY DATA ANALYSIS

### Renaming Column

```
data.rename(columns = {"Unnamed: 0": "Date"}, inplace = True)
data.columns
```

### Removing Un-authentic Columns and Rows

```
data.drop(["Estonian kroon", "Lithuanian litas", "Latvian lats", "Slovak koruna"], axis = 1, inplace = True)
data = data[1:]

data.head()
```

### Checking Empty Values

```
if data.isna().sum().sum() ==0:
  print("NO EMPTY VALUES!!!")
```

### VISUALIZATION

```
data = data.melt(id_vars=["Date"], var_name = "Currency", value_name = "ExchangeRate")
data.sort_values(["Date"], inplace = True)
print("Dimesnion: ", data.shape)
data.head()
```

```
x = data[['ExchangeRate']].values.astype(float)
data[['ExchangeRate']] = pd.DataFrame(preprocessing.MinMaxScaler().fit_transform(x))
```

```
fig = px.bar(data, x='Currency', y="ExchangeRate", animation_frame = "Date", color = "ExchangeRate")
fig.show()
```

```
fig = px.area(data, x='Date', y="ExchangeRate", color="Currency", line_group="ExchangeRate")
fig.show()
```
