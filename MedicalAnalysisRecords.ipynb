### IMPORT LIBRARIES

```
import pandas as pd
import numpy as np
import plotly.express as px

print("LIBARARIES IMPORTED!!!")
```

### IMPORT DATA

```
path = "/content/drive/My Drive/Final Project/Project Data/Medical Analysis/CovidStatisticsProfileHPSCIrelandOpenData.csv"

data = pd.read_csv(path)
print("Dimensions of Data: ", data.shape)
```

```
data.head()
```

##EXPLORATORY DATA ANALYSIS

### Check Empty Values

```
data = data.dropna()

print("EMPTY VALUES: ", data.isna().sum().sum())
# print("NEW DIMENSIONS: ", data.shape)
```

## HOSPITALISED

### Create New Dataset

```
hospitalised = data[['Date', 'HospitalisedAged5', 'HospitalisedAged5to14', 'HospitalisedAged15to24', 'HospitalisedAged25to34', 'HospitalisedAged35to44', 'HospitalisedAged45to54',
       'HospitalisedAged55to64', 'HospitalisedAged65up']] 
hospitalised.head()
```

```
hospitalised = hospitalised.melt(id_vars=["Date"], var_name="HospitalisedGroups", value_name="PatientCount")
hospitalised.sort_values(["Date", "HospitalisedGroups"], inplace = True) 
hospitalised.head()
```

```
fig = px.bar(hospitalised, x='HospitalisedGroups', y="PatientCount", 
                 animation_frame = "Date",
                 color = "PatientCount")
fig.show()
```

## HOSPITALISED

```
age = data[["Date", "Aged1",	"Aged1to4",	"Aged5to14",	"Aged15to24",	"Aged25to34",	"Aged35to44",	"Aged45to54",	"Aged55to64",	"Aged65up",	"Median_Age"]]
age.head()
```

```
age = age.melt(id_vars=["Date"], var_name="AgeGroup", value_name="PatientCount")
age.sort_values(["Date", "AgeGroup"], inplace = True) 
age.head()
```

```
fig = px.bar(age, x='AgeGroup', y="PatientCount", 
                 animation_frame = "Date", animation_group = "AgeGroup",
                 color = "PatientCount")
fig.show()
```

```
fig = px.bar(traveltransmission, x='TransmissionReason', y="PatientCount", 
                 animation_frame = "Date", animation_group = "TransmissionReason",
                 color = "TransmissionReason")
fig.show()
```
