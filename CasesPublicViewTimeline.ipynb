### IMPORT LIBRARIES

```
import pandas as pd
import numpy as np
import plotly.express as px
from datetime import datetime as dt

print("LIBARARIES IMPORTED!!!")
```

### IMPORT DATA

```
path = "/content/drive/My Drive/Final Project/Project Data/Cases/ICUBISHistoricTimelinePublicView.csv"

data = pd.read_csv(path).drop("OBJECTID", axis = 1)
print("Dimensions of Data: ", data.shape)
```

```
data.head()
```

## EXPLORATORY DATA ANALYSIS

### Renaming Column Names

```
#data["extract"] = pd.to_datetime(data["extract"]).dt.date
data.rename({"extract": "Date", "ncovidconf":"ConfirmCases",	"ndischcovidconf":	"DischargedCases", "adcconf": "AdmitCases"}, axis = 1, inplace = True)

data.head()
```

### Checking Empty Values

```
data.isna().sum()
```

```
data["DischargedCases"] = data["DischargedCases"].fillna(0)

print("EMPTY VALUES: ", data.isna().sum().sum())
# print("NEW DIMENSIONS: ", data.shape)
```

### Changing Data Type to "int"

```
data["DischargedCases"] = data["DischargedCases"].astype(int)
data.head()
```

### Reframing Columns to Categorical Rows

```
data = data.melt(id_vars=["Date"], var_name="Cases", value_name="Count")
data.sort_values(["Date"], inplace = True)
data.head()
```

## VISUALIZATION

```
fig = px.bar(data, x='Count', y="Cases", color = "Cases", animation_frame = "Date", orientation="h")
fig.show()
```

```
fig = px.area(data, x='Cases', y="Count", animation_frame = "Date", line_group = "Count", color = "Count")
fig.show()
```
