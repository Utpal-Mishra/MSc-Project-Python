### IMPORT LIBRARIES

```
import pandas as pd
import numpy as np
import plotly.express as px
from datetime import datetime as dt
from sklearn import preprocessing

print("LIBARARIES IMPORTED!!!")
```

### IMPORT DATA

```
path = "/content/drive/MyDrive/Final Project/unemployment_by_country_by_year.csv"

data = pd.read_csv(path, skiprows = 4)
print("Dimensions of Data: ", data.shape)
```

```
data.sort_values(['2020'], ascending=False)
```

## EXPLORATORY DATA ANALYSIS

### Removing Un-authentic Columns and Rows

```
data.drop(["Indicator Name", "Indicator Code"], axis = 1, inplace = True)
data.shape
```

### Checking Empty Values

```
if data.isna().sum().sum() ==0:
  print("NO EMPTY VALUES!!!")
else:
  print("WARNING: EMPTY VALUES!!!")
```

```
data.isna().sum()
```

```
data[["Country Name", "Country Code"]]
```

```
data  = data[["Country Name", "Country Code"]].merge(data.drop(["Country Name", "Country Code"], axis = 1).dropna(how = "all"), left_index=True, right_index=True, how='inner')
print("NEW DIMENSIONS: ", data.shape)
```

### VISUALIZATION

```
data = data.melt(id_vars=["Country Name", "Country Code"], var_name = "Year", value_name = "UnemploymentRate")
#data.sort_values(["Date"], inplace = True)
print("Dimesnion: ", data.shape)
data.head()
```

```
fig = px.bar(data, x='Country Name', y = "UnemploymentRate", animation_frame = "Year", color = "UnemploymentRate")
fig.show()
```

```
fig = px.area(data, x='Country Name', y = "UnemploymentRate", color = "UnemploymentRate", line_group="Year")
fig.show()
```
